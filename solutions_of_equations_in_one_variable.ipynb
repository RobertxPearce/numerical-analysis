## Bisection Method
Based on the **Intermediate Value Theorem**. The method repeatedly bisects subintervals of $[a,b]$ and at each step locates the half containing p with $f(p)=0$.

This method is simple to implement and will always converge to a solution but is not efficient and relatively slow to converge.

Example: Find a solution to $f(x)=x^3-3$ on the interval $[1,3]$.

```python
def bisection_method(a, b, tol, n):
    """
    :param a: Left endpoint
    :param b: Right endpoint
    :param tol: Tolerance
    :param n: Maximum number of iterations
    :return: Root or solution of an equation of the form f(x)=0
    """
    i = 1
    fa = a**3-3
    fb = b**3-3

    if fa * fb > 0:
        print("Intermediate Value Theorem does not hold: f(a) * f(b) > 0")

    while i < n:
        p = a + (b-a)/2
        fp = p**3-3
        if fp == 0 or (b-a)/2 < tol:
            print(f"Solution after {i} iterations: {p}")
            break
        i = i + 1
        if fa * fp > 0:
            a = p
            fa = fp
        else:
            b = p
        if i == n:
            print(f"Method filed after {i} iterations")

bisection_method(1, 3, 0.0001, 20)
```

```python
def bisection_graph(x):
    return x**3-3

x = np.linspace(0, 3, 100)
y = bisection_graph(x)

plt.title("Bisection Method Example")
plt.grid(True)
plt.axhline(0, color='black')
plt.axvline(0, color='black')
plt.plot(x, y, label=r"$f(x) = x^3 - 3$")
plt.scatter(1.44219970703125, 0, color='red', label=r"$f(x)=0$")
plt.legend()
```
